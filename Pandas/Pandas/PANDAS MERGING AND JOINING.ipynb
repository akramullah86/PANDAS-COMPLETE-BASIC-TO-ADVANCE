```python
# 9. find top 3 students who did most number enrollments
```

```python
(
    students.merge(registrations,how='inner',on='student_id')
    .merge(courses,how='inner',on='course_id')
    .groupby(['student_id','name'])
    .name.count()
    .sort_values(ascending=False)
    .head(3)
)
```

```python
# another methodd to solev the above question
(
    students.merge(registrations,how='inner',on='student_id')
    .merge(courses,how='inner',on='course_id')
    [['student_id','name']]
    .value_counts()
    .sort_values(ascending=False)
    .head(3)


)
```

```python
# 10. find top 3 students who spent most amount of money on courses
(
    students.merge(registrations,how='inner',on='student_id')
    .merge(courses,how='inner',on='course_id')
    .groupby('name')
    ['price']
    .sum()
    .sort_values(ascending=False)
    .head(3)


)
```
