```python
# 5. find students who enrolled in both the months
reg_student = np.intersect1d(nov['student_id'],dec['student_id'])
students[students['student_id'].isin(reg_student)]
```

```python
# 6. find course that got no enrollment
not_student = np.setdiff1d(courses['course_id'],reg['course_id'])

courses[courses['course_id'].isin(not_student)]
```

```python
# 7. find students who did not enroll into any courses
new = students.merge(reg,how='inner',on='student_id')
new1 = np.setdiff1d(courses['course_id'],new['course_id'])
students[students['student_id'].isin(new1)]
```

```python
st = np.setdiff1d(students['student_id'],reg['student_id'])
students[students['student_id'].isin(st)]
```

```python
# 9. find top 3 students who did most number enrollments
students.merge(reg,how='inner',on='student_id')[['student_id','name']].value_counts().sort_values(ascending=False).head(3)
# reg
```

```python
# 10. find top 3 students who spent most amount of money on courses
students.merge(reg,how='inner',on='student_id').merge(courses,how='inner',on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
# courses
```
